# Import Dependencies

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

# Import Dataset

```python
df = pd.read_csv('/content/Salary_Data.csv')
```

# Split Variables

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

# Split Dataset Train - Test

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 1/3, random_state = 0)
```

# Create Model Linear Regression

```python
rls = LinearRegression()
rls.fit(X_train, y_train)
```

# Test Prediction

```python
y_pred = rls.predict(X_test)
```

# Visualization Train Result

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_train, y_train, color="red", label="Training Data", edgecolors='k')
plt.plot(X_train, rls.predict(X_train), color="blue", label="Regression Model")
plt.title("Salary vs Years of Experience (Training Set)", fontsize=14)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary ($)", fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(loc='best')
plt.show()
```

# Visualization Test Result

```python
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color="red", label="Test Data", edgecolors='k')
plt.plot(X_train, rls.predict(X_train), color="blue", label="Regression Model")
plt.title("Salary vs Years of Experience (Test Set)", fontsize=14)
plt.xlabel("Years of Experience", fontsize=12)
plt.ylabel("Salary ($)", fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend(loc='best')
plt.show()
```
