```python
import nltk
import re
```

```python
nltk.download('cess_esp')
```

```python
corpus = nltk.corpus.cess_esp.sents()
print(corpus)
```

```python
print(len(corpus))
```

```python
flatten = [w for l in corpus for w in l]
print(flatten[:20])
```

```python
array = [w for w in flatten if re.search('es', w)]
print(array)
```

```python
array = [w for w in flatten if re.search('es$', w)] # Expresion regular para que el meta caracter este al final de la palabra.
print(array)
```

```python
array = [w for w in flatten if re.search('^[het]', w)] # Expresion regular para que el meta caracter este al inicio de la palabra dependiendo el rango [...].
print(array)
```

```python
# Clausura * 0 veces, + 1 o mas veces
array = [w for w in flatten if re.search('^se+', w)]
print(array)
```
