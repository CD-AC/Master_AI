## Restricciones Regresi√≥n Lineal Multiple
1. Linealidad: Se asume que la relaci√≥n entre las variables independientes y la variable dependiente puede describirse de forma lineal. Esto significa que un cambio en la variable independiente produce un efecto constante en la variable dependiente.

2. Homocedasticidad: Implica que la varianza de los errores (o residuos) es constante a lo largo de todos los valores de las variables independientes. En otras palabras, se espera que la dispersi√≥n de los errores sea la misma sin importar el nivel de las variables predictoras.

3. Normalidad multivariable: Se requiere que los errores de la regresi√≥n (o, en un contexto m√°s amplio, los valores de las variables) sigan una distribuci√≥n normal. Esto ayuda a que los intervalos de confianza y las pruebas de hip√≥tesis sean v√°lidas.

4. Independencia de los errores: Establece que los errores no deben estar correlacionados entre s√≠. Por ejemplo, en datos de series temporales, los errores de un per√≠odo no deber√≠an influir en los errores de otros per√≠odos.

5. Ausencia de multicolinealidad: Significa que las variables independientes no deben estar fuertemente correlacionadas entre s√≠. Una alta correlaci√≥n entre variables independientes puede distorsionar las estimaciones de los coeficientes y dificultar la interpretaci√≥n del modelo.

## p-valor
Es la probabilidad, suponiendo que la hip√≥tesis nula sea verdadera, de obtener un resultado tan extremo (o m√°s extremo) que el observado en tu muestra de datos. En otras palabras, mide la evidencia en contra de la hip√≥tesis nula: mientras m√°s peque√±o sea el p-valor, m√°s evidencia hay para rechazar dicha hip√≥tesis. Si el p-valor es menor que un nivel de significancia establecido (por ejemplo, 0.05), normalmente se concluye que existe suficiente evidencia estad√≠stica para descartar la hip√≥tesis nula.
